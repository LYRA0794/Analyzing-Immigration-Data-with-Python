## Set the 'Country' column as the index

This part is must to do because the default index of the dataset is a numeric range from 0 to 194. This makes it very difficult to do a query by a specific country. For example to search for data on Japan, we need to know the corresponding index value.

```python
df.set_index('Country', inplace=True)
```

```python
df.head(3)
```

## Convert the column names into strings: '1980' to '2013'

Column names that are integers (such as the years) might introduce some
confusion. For example, when we are referencing the year 2013, one might confuse that when the 2013th positional index. To avoid this ambuigity, let's do the convert part.


```python
df.columns = list(map(str, df.columns))
for x in df.columns.values:
   print(type(x))
```

```python
a = df.columns.values
print(a)
```

Since we converted the years to string, let's declare a variable that will allow us to easily call upon the full range of years:

```python
years = list(map(str, range(1980, 2014)))
years
```

# Visualizing Data using Matplotlib

Step 2: Plot the graph. We will explicitly specify a line plot by passing the `kind` parameter to `plot()`.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df_CI.plot(kind='line')
```

That doesn't look right, *pandas* should plot indexes on the x-axis and columns as individual lines on the y-axis. Since `df_CI` is a dataframe with `Country` as the index and `year` as the column, we must first transform the dataframe using the `transpose()` method to swap rows and columns.

```python
df_CI_new = df_CI.transpose()
df_CI_new.head()
```

```python
df_CI_new.index = df_CI_new.index.map(int)
df_CI_new.plot(kind='line')

plt.title('Immigrants from China and India')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

<br>From the above plot, we can observe that the China and India have very similar immigration trends through the years.
